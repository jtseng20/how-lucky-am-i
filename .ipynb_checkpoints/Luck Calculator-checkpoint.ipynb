```python
import random
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')

# How important is luck?
# Under this model, an "applicant" pool of size SIZE is composed of entrants whose aptitudes are modeled as a tuple of
# uniform quantities skill and luck, whose ratio is set by default to 95:5. Entrants with the highest summed scores are 
# "admitted". The Skill and Luck distribution are plotted, and the question "what if Luck were not a factor?" is answered in
# this simulation by counting which admits were also in the respective top portion of the population in Skill alone.
# As expected, when the admit rate is extremely low ( < 0.001%), the Luck of admits is sharply skewed (although of course, 
# they are also extremely Skilled), and only about 10% would have made the cut with skill alone.

SIZE = 40000
ADMITRATE = 0.04
ADMITS = int(ADMITRATE * SIZE)

LUCK = 0.05
SKILL = 1 - LUCK

def randomTuple():
    return ((random.random()) * SKILL, (random.random()) * LUCK)


entrants = [randomTuple() for i in range(SIZE)]

admittedCombined = sorted(entrants, key = lambda e : e[0] + e[1])[-ADMITS:]
allLuckOnly = sorted(entrants, key = lambda e : e[1])
admittedLuckOnly = allLuckOnly[-ADMITS:]
admittedSkillOnly = sorted(entrants, key = lambda e : e[0])[-ADMITS:]

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

x = [a[1] for a in entrants]
y = [a[1] for a in admittedCombined]
ax[0].hist([x, y], density = True, bins = 100, label = ["All Luck", "Admit Luck"])
ax[0].legend(loc='upper left')

x = [a[0] for a in entrants]
y = [a[0] for a in admittedCombined]
ax[1].hist([x, y], density = True, bins = 100, label = ["All Skill", "Admit Skill"])

skillFloor = admittedSkillOnly[0][0]
ax[1].axvline(x = skillFloor, color = "red", linestyle = "dashed", label = "Admit Threshold (Skill Only)")
ax[1].legend(loc='upper left')

combinedTopSkillCount = 0
for i in range(ADMITS):
    combinedTopSkillCount += (admittedCombined[i][0] >= skillFloor)
fig.suptitle("With a " + str(ADMITRATE * 100) + "% admit rate, "+ str(combinedTopSkillCount / ADMITS * 100) + "% of admits would have been admitted with skill alone", size = "xx-large");
plt.show()
```
